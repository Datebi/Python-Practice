```python
#1 Importo librerias necesarias
import pandas as pd
import numpy as np
import requests
```

```python
#2. ETL de datos
# Usuarios a agregar
# Lista de Usuarios 1 
Usuarios_0 = 'users'

# URLs pre-determinadas por la documentacion de JsonPlaceHolder
# Para las listas de usuarios:
Url_pre = 'https://jsonplaceholder.typicode.com/'

```

```python
3# Genero un array con los usuarios
array_lists =    np.array([ Url_pre + Usuarios_0])
```

```python
# Definimos el df
df_usuarios = pd.DataFrame()
for list in array_lists:
  
    # Ejecutamos peticion para traer lista
    response_json = requests.get( list )
        # Verificamos primero si la peticion se realizo correctamente
    if response_json.status_code == 200:
        response_json = response_json.json()
        print('--CARGANDO LISTA DE USUARIOS--')

        # Recorro las listas de usuarios
        for user in response_json:

            # Guardamos la informacion
            id                     = user['id']
            name                   = user['name']
            username               = user['username']
            email                  = user['email']
            street                 = user['address']['street']
            suite                  = user['address']['suite']
            city                   = user['address']['city']
            zipcode                = user['address']['zipcode']
            lat                    = user['address']['geo']['lat']
            lng                    = user['address']['geo']['lng']
            phone                  = user['phone']
            website                = user['website']
            cpyname                = user['company']['name']
            catchPhrase            = user['company']['catchPhrase']
            bs                     = user['company']['bs']



                # Guardamos en el df la informacion
            df_usuarios = df_usuarios.append({'Id'                  : id ,
                                           'name'                   : name ,
                                           'username'               : username ,
                                           'email'                  : email ,
                                           'street'                 : street ,
                                           'suite'                  : suite ,
                                           'city'                   : city ,
                                           'zipcode'                : zipcode ,
                                           'lat'                    : lat ,
                                           'lng'                    : lng ,
                                           'phone'                  : phone ,
                                           'website'                : website ,
                                           'cpyname'                : cpyname ,
                                           'catchPhrase'            : catchPhrase ,
                                           'bs'                     : bs }, ignore_index=True)                                          
    else:
      print('Error de peticion.')
      print( response_json.status_code )
                    
```
