>> Consigna: Leer los set de datos w_data.csv y p_data.csv, fusionarlos a partir de la variable que tengan en común (ID)y realizar las siguientes ejercicios:
1. Cuantos ID tienen en común. Nombrar cuales no tienen en común
2. Corregir los target a solo dos valores distintos ‘<=50k’ y ‘>50k’
3. Calcular la proporción que hay entre personas con salarios >50k respecto del total para entre personas de distinto sexo, Raza y Educacion. Por ejemplo: Cuál es el porcentaje de mujeres con un salario mayor a 50K y compararlo con el de los hombres. Lo mismo para personas de distintas razas y Educación.

```python
# 1. Importo las librerias necesarias
import pandas as pd
import csv
```

```python
# 2. Carga del archivo CSV


df_tarea_baigorria_p = pd.read_csv('p_data.csv' ,sep = ';', encoding='latin-1')
df_tarea_baigorria_w = pd.read_csv('w_data.csv' ,sep = ';', encoding='latin-1')

#print(df_tarea_baigorria)
df_tarea_baigorria_p.describe(percentiles=None, include=None, exclude=None, datetime_is_numeric=False)
df_tarea_baigorria_w.describe(percentiles=None, include=None, exclude=None, datetime_is_numeric=False)
```

```python
#Seteamoos esta configuracion para poder ver que todas las columnas se hayan cargado correctamente
pd.set_option('display.max_columns', 15)
#Imprimo el DF
print(df_tarea_baigorria_p)
print(df_tarea_baigorria_w)
```

```python
#Busco cuantos IDS tiene en comun W en P, ya que P contiene la menor cantidad de registros ID
#Arrojando como resultado 48842 valores en comun unicos.

df_idscomun = df_tarea_baigorria_p[df_tarea_baigorria_p['ID'].isin(df_tarea_baigorria_w['ID'])]
print(df_idscomun.nunique())
```

```python
#Busco cuantos IDS NO tiene en comun para mostrarlos.
#Dando como resultado 1 id: 1344995623   

df_idsnotcomun = df_tarea_baigorria_w[~df_tarea_baigorria_w['ID'].isin(df_tarea_baigorria_p['ID'])]
print(df_idsnotcomun.nunique())
print(df_idsnotcomun)
```

```python
#Practico el merge

df_fusion = pd.merge(df_tarea_baigorria_p, df_tarea_baigorria_w, on="ID")
df_fusion

```

```python
#Practicamos la correccion de la columna Target
#df_fusion['Target'] = df_fusion['Target'].astype('str') 
df_fusion['Target'] = df_fusion['Target'].replace({'<=50K.': '<=50K', '>50K.': '>50K'}, regex= True)
```

```python
print(df_fusion.nunique())
#Se corrobora que hay 2 valores unicos de target (Previo a la correccion figuraban 4)
```

```python
df_corroboracion =  df_fusion.groupby(['Target','Sex']).agg({'Target': 'count'})
print(df_corroboracion)
#Se corrobora que son los valores unicos que queriamos con la correccion realizada.
```

```python
#Genero los DF segun las condiciones solicitadas, salarios mayores a 50k y segmentado por:

#Sexo
df_50khvsm =  df_fusion.groupby(['Target','Sex']).agg({'Target': 'count'})
df_porcentajehvsm = df_50khvsm.groupby(level=0).apply(lambda x:100 * x / float(x.sum()))

#Raza
df_50krvsr =  df_fusion.groupby(['Target','Race']).agg({'Target': 'count'})
df_porcentajervsr = df_50krvsr.groupby(level=0).apply(lambda x:100 * x / float(x.sum()))
#Educacion
df_50kevse =  df_fusion.groupby(['Target','Education']).agg({'Target': 'count'})
df_porcentajeevse = df_50kevse.groupby(level=0).apply(lambda x:100 * x / float(x.sum()))

print(df_porcentajehvsm)
print(df_porcentajervsr)
print(df_porcentajeevse)
```
